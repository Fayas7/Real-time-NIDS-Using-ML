EXPLARATORY DATA ANALYSIS

# Import necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load the preprocessed training data

```python
train_data = pd.read_csv(r'C:\Users\fayaz\Documents\NIDS_IMPLEMENTATION\data\X_train_processed.csv')
labels = pd.read_csv(r'C:\Users\fayaz\Documents\NIDS_IMPLEMENTATION\data\y_train_processed.csv')
```

```python
# Combine features and labels for EDA
train_data['Label'] = labels
```

```python
# Display the first few rows of the dataset
train_data.head()
```

```python
# Check for missing values
missing_values = train_data.isnull().sum()
print("Missing values in training data:\n", missing_values[missing_values > 0])
```

```python

# Summary statistics
summary_statistics = train_data.describe()
print("Summary statistics:\n", summary_statistics)
```

```python
# Visualize the distribution of the target variable
plt.figure(figsize=(8, 5))
sns.countplot(x='Label', data=train_data)
plt.title('Distribution of Target Variable')
plt.xlabel('Label')
plt.ylabel('Count')
plt.show()
```

```python
# Visualize the distribution of numerical features
numerical_cols = train_data.select_dtypes(include=['float64', 'int64']).columns
```

```python
plt.figure(figsize=(20, 15))  # Increase the figure size
num_cols = len(numerical_cols)
rows = (num_cols // 4) + (num_cols % 4 > 0)  # Calculate the number of rows needed

for i, col in enumerate(numerical_cols):
    plt.subplot(rows, 4, i + 1)  # Adjust the number of rows as needed
    sns.histplot(train_data[col], bins=30, kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)  # Add x-label for clarity
    plt.ylabel('Frequency')  # Add y-label for clarity

# Adjust layout
plt.subplots_adjust(hspace=0.4, wspace=0.4)  # Adjust space between subplots
plt.tight_layout()  # Apply tight layout
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(12, 8))

# Select only numeric columns for correlation
numeric_data = train_data.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
correlation_matrix = numeric_data.corr()

# Create the heatmap
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Correlation Heatmap')
plt.show()
```
